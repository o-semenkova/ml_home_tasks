1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = 10*x - 5
derivative_f = sp.diff(f, x)
print(derivative_f)
```

```python
# Створення графіку

x = np.linspace(0, 10, 10)

# Бо наша похiдна це константа
y_df = np.full(10, 10)

f = 10*x-5

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, y_df, label="f'(x) = 10", color='green')

# Позначення точки та дотичної лінії
x_point = 3
f_point = x_point*10 - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='purple') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3') # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функція залежності оцінки від кількості годин, витрачених на навчання")
plt.xlabel("Години")
plt.ylabel("Оцiнка")
plt.legend()
plt.grid(True)

plt.show()
```

**Спостереження**

Коли тангенс лінія співпадає з функцією, а похідна є константою, це означає, що ми маємо справу з лінійною функцією. Детальніше:

 - Лінійна функція:

Загальний вигляд: f(x) = mx + b, де m - нахил, b - зсув по осі y

 - Тангенс лінія:

Для лінійної функції тангенс лінія в будь-якій точці співпадає з самою функцією

 - Похідна:

Похідна лінійної функції f(x) = mx + b є константою: f'(x) = m

 - Геометрична інтерпретація:

Графік функції - пряма лінія
Нахил цієї прямої дорівнює значенню похідної (константі m)

 - Властивості:

  - Постійна швидкість зміни (однаковий приріст y для будь-якого приросту x)
  - Графік не має вигинів чи екстремумів

 - Значення в математиці та фізиці:

  - Описує рівномірний рух в фізиці
  - Є найпростішим видом неперервної функції

 - Застосування:

  - Моделювання простих лінійних залежностей
  - Апроксимація складніших функцій на малих інтервалах

**Висновок:**
 - Отже, коли тангенс лінія співпадає з функцією, а похідна є константою, ми маємо справу з лінійною функцією, яка графічно представляє собою пряму лінію. Це найпростіший вид неперервної функції, яка має широке застосування в математиці, фізиці та інших галузях науки.

 - Змiна кiлькостi годин прямо пропорцiонально впливае на оцiнку: чим бiльше вчимось, тим бiльшою буде оцiнка

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.log(x, 10)


derivative_f1 = sp.diff(f1, x)
print(derivative_f1)

derivative_f2 = sp.diff(f2, x)
print(derivative_f2)
```

```python
x = np.linspace(-30, 10, 100)

f1 = np.sin(x)
f1_derivative = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f1, label='f(x) = sin(x)', color='orange')
plt.plot(x, f1_derivative, label="f'(x) = cos(x)", color='blue')

plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
x = np.linspace(-300, 300, 100)

f2 = np.log10(x)
f2_derivative = 1/(x*np.log(10))

plt.figure(figsize=(10, 5))
plt.plot(x, f2, label='f(x) = ln(x)', color='green')
plt.plot(x, f2_derivative, label="f'(x) = 1/(x*log(10))", color='red')

plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

f = x**2 - 5*x + 1

derivative_f = sp.diff(f, x)
print(derivative_f)
```

```python
x = np.linspace(0, 100, 500)

f = x**2 - 5*x + 1
f_derivative = 2*x - 5

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x^2 - 5x + 1', color='purple')
plt.plot(x, f_derivative, label="f'(x) = 2x - 5", color='yellow')

plt.title("Графік функції загальних витрат компанії від обсягу виробництва та її похідної")
plt.xlabel("Кiлькiсть вироблених одиниць")
plt.ylabel("Витрати компанii")
plt.legend()
plt.grid(True)

plt.show()
```
